```python
import cv2
from img_dir_scanner import img_dir_scanner
import os
```

```python
def save_video(directory, filename, image_container, min_images=4):
    if len(image_container) > min_images:
        vidname = os.path.join(directory, filename + ".avi")

        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        vid_writer = cv2.VideoWriter(vidname,fourcc, 10, (image_container[0].shape[1],image_container[0].shape[0]))

        for image in image_container:
            vid_writer.write(image)

        vid_writer.release()
        
def image2Vid(directory):
    
    imgSource = img_dir_scanner(directory)
    num_images = len(imgSource.image_list)
    images_left = num_images
    image_container =[]
    image_container.append(cv2.imread(imgSource.image_list[0]))

    filename = imgSource.basenames_wo_ext[0]
    print("processing ", directory)
    for i in range(1, imgSource.image_number):
        cur_filename = imgSource.basenames_wo_ext[i]
        timediff = int(cur_filename) - int(filename)

        if timediff > 10:    
            save_video(directory, cur_filename, image_container)
            image_container = []

        image_container.append(cv2.imread(imgSource.image_list[i]))    
        filename = cur_filename
       
    print("done")
```

```python
dir_list = ['/media/newhd5/cam_photos/0806',
           '/media/newhd5/cam_photos/0807','/media/newhd5/cam_photos/0808',
           '/media/newhd5/cam_photos/0809']

for d in dir_list:
    image2Vid(d)
     
```
